### Fill the NA values by the mean of it

```python
df = df.fillna(df.mean()["2022 Population":"Growth Rate"])
df.isnull().sum()
```

### Group tables by every continent existed

```python
df_con = df.groupby("Continent")[['1970 Population','1980 Population','1990 Population',
'2000 Population',
'2010 Population',
'2015 Population',
'2020 Population',
'2022 Population']].mean().sort_values(by='2022 Population', ascending= False)
df_con
```

```python
df.corr()
```

### Checking correlation of every column existed in the table

```python
sns.heatmap(df.corr(), annot=True)
plt.rcParams['figure.figsize'] = (15,7)

plt.show()
```

```python
df_con2 = df_con.transpose()
df_con2
```
